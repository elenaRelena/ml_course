```python
import numpy as np
import pandas as pd
import sklearn
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'Admission_Predict.csv', sep=",")
data.head()
```

```python
data.shape
```

```python
for col in data.columns:
    # Количество пустых значений - все значения заполнены
    temp_null_count = data[data[col].isnull()].shape[0]
    print('{} - {}'.format(col, temp_null_count))
```

```python
ax = sns.barplot(x="TOEFL Score", y="Chance of Admit ", data=data)
```

```python
ax = sns.barplot(x="GRE Score", y="Chance of Admit ", data=data)
```

```python
sns.boxplot('University Rating','TOEFL Score', data = data)
```

```python
corrmat = data.corr()
fig,ax = plt.subplots(figsize = (12,9))
sns.heatmap(corrmat, annot=True, vmax=.8, square=True)
corrmat
```

```python
sns.distplot(data['LOR '])
```

```python
y = np.array([data["TOEFL Score"].min(),data["TOEFL Score"].mean(),data["TOEFL Score"].max()])
x = ["Worst","Average","Best"]
plt.bar(x,y)
plt.title("TOEFL Scores")
plt.xlabel("Level")
plt.ylabel("TOEFL Score")
plt.show()
```

```python
data["GRE Score"].plot(kind = 'hist',bins = 200,figsize = (6,6))
plt.xlabel("GRE Score")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.scatter(data["GRE Score"],data.CGPA)
plt.xlabel("GRE Score")
plt.ylabel("CGPA")
plt.show()
```

```python
s = data[data["Chance of Admit "] >= 0.75]["University Rating"].value_counts().head(5)
s.plot(kind='bar',figsize=(20, 10))
plt.show()
```

```python
fig = sns.distplot(data['GRE Score'], kde=True)
plt.show()

fig = sns.distplot(data['TOEFL Score'], kde=True)
plt.show()

fig = sns.distplot(data['University Rating'], kde=True)
plt.show()

fig = sns.distplot(data['SOP'], kde=True)
plt.show()

fig = sns.distplot(data['CGPA'], kde=True)
plt.show()

plt.show()
```
